```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
base = pd.read_csv("airpassengers.csv")
```

```python
dateparse = lambda dates: pd.datetime.strptime(dates, "%Y-%m")
```

```python
base = pd.read_csv("airpassengers.csv", parse_dates=["Month"], index_col="Month", date_parser=dateparse)
```

```python
ts = base["#Passengers"]
ts.head()
```

```python
plt.plot(ts)
```

```python
decomposicao = seasonal_decompose(ts)
tendencia = decomposicao.trend
sazonal = decomposicao.seasonal
aleatorio = decomposicao.resid
```

```python
plt.plot(tendencia)
```

```python
plt.plot(sazonal)
```

```python
plt.plot(aleatorio)
```

```python
plt.subplot(4,1,1)
plt.plot(ts, label="Original", c="black")
plt.legend(loc="best")

plt.subplot(4,1,2)
plt.plot(tendencia, label="Tendncia", c="blue")
plt.legend(loc="best")

plt.subplot(4,1,3)
plt.plot(sazonal, label="Sazonalidade", c="orange")
plt.legend(loc="best")

plt.subplot(4,1,4)
plt.plot(aleatorio, label="Aleatrio", c="red")
plt.legend(loc="best")

plt.tight_layout()
```
